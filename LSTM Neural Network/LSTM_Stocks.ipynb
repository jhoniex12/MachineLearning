# COIT29225 Assessment 3 – LSTM Stock Price Forecasting (BHP, CBA, WES)


```python
#Import Libraries

import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split

from tensorflow.keras.callbacks import EarlyStopping

# Global Variables
PERFORMANCE_LOG = []
STOCK_LIST = ["BHP", "CBA", "WES"]
RESULT_PATH = "./lstm_result/"

os.makedirs(RESULT_PATH, exist_ok=True)#create folder if missing

#Load Dataset
def load_and_preprocess_data(stock_name, set_neurons=100, set_epochs=50):
    df = pd.read_csv(f"{stock_name}.csv")
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")

    df["ClosePrice"] = (
        df["ClosePrice"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.strip()
        .astype(float)
    )
    prices = df["ClosePrice"].values.reshape(-1, 1)

    lookback = 30

    X = []
    y = []

    for i in range(lookback, len(prices)):
        X.append(prices[i - lookback:i])
        y.append(prices[i])

    X = np.array(X)
    y = np.array(y)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        train_size=0.7,
        test_size=0.3,
        shuffle=False
    )

    scaler = MinMaxScaler()
    y_train_scaled = scaler.fit_transform(y_train)
    y_test_scaled = scaler.transform(y_test)
    X_train_scaled = scaler.transform(X_train.reshape(-1, 1)).reshape(X_train.shape)
    X_test_scaled = scaler.transform(X_test.reshape(-1, 1)).reshape(X_test.shape)

    neurons = set_neurons
    epochs_used = set_epochs

    model = Sequential([
        LSTM(neurons, input_shape=(lookback, 1)),
        Dense(1)
    ])

    model.compile(optimizer="adam", loss="mse")

    early_stop = EarlyStopping(
        monitor="val_loss",
        patience=10,
        restore_best_weights=True
    )

    history = model.fit(
        X_train_scaled, y_train_scaled,
        epochs=epochs_used,
        batch_size=32,
        validation_data=(X_test_scaled, y_test_scaled),
        callbacks=[early_stop]
    )

    # Prediction and Evaluation
    predicted = model.predict(X_test_scaled)

    y_test_real = scaler.inverse_transform(y_test_scaled)
    predicted_real = scaler.inverse_transform(predicted)

    # Calculate RMSE and R2
    rmse = np.sqrt(mean_squared_error(y_test_real, predicted_real))
    print("RMSE:", rmse)
    r2 = r2_score(y_test_real, predicted_real)
    print("R2 Score:", r2)

    # Visualization plot
    plt.figure(figsize=(10, 5))
    plt.plot(y_test_real, label="Actual Price")
    plt.plot(predicted_real, label="Predicted Price")
    plt.title(f"{stock_name} Stock Price Prediction Neurons: {neurons} | Epoch: {epochs_used} \n RMSE:{rmse:.2f} R2: {r2:.2f}")
    plt.xlabel("Days")
    plt.ylabel("Close Price")
    plt.legend()
    plt.savefig(f"{RESULT_PATH}{stock_name}_{neurons}_{epochs_used}.png")
    plt.show()
    
    PERFORMANCE_LOG.append ( f"{stock_name} Stock Price Prediction Neurons: {neurons} | Epoch: {epochs_used} | RMSE:{rmse:.2f} | R2: {r2:.2f}")

def save_log(stock_name):
    with open(f"{RESULT_PATH}{stock_name}_performance_log.txt", "w") as f:
        for item in PERFORMANCE_LOG:
            print(item)
            f.write(str(item) + "\n")


def run_test():
    for stock_name in STOCK_LIST:
        load_and_preprocess_data(stock_name, set_neurons=20, set_epochs=20)
        load_and_preprocess_data(stock_name, set_neurons=50, set_epochs=20)
        load_and_preprocess_data(stock_name, set_neurons=100, set_epochs=20)
        load_and_preprocess_data(stock_name, set_neurons=20, set_epochs=50)
        load_and_preprocess_data(stock_name, set_neurons=50, set_epochs=50)
        load_and_preprocess_data(stock_name, set_neurons=100, set_epochs=50)
        save_log(stock_name)

run_test()

```
